# Let's start with the spring equation

# $$ m \ddot{x} - \nu \dot{x} + \kappa_0 x = 0 $$

# Question 1: What does each term in the equation mean physically?

# A: The first term is mass multiplied by acceleration of the mass. The next two terms are the forces applied to the mass given the position of the mass. The term $-\nu \dot{x}$ represents a friction force, where energy is dissipated from the system as the mass moves. The term $\kappa_0 x$ is the restoring force of the spring. This is based on Hooke's law.  

# Question 2: How can we convert this to a system of first order ODEs?

# A: We can convert the system by setting a dummy variable

# $$ y = \dot{x} $$

# This equation is our first equation in the next system. The equation also implies that $\dot{y} = \ddot{x}$. Therefore, the second equation in our system is the original ODE rewritten in terms of $x$, $y$, and $\dot{y}$, 


# $$ m \dot{y} - \nu y + \kappa_0 x = 0 $$

# The system is written in the more typical form as, 

# $$ \dot{x} = y $$
# $$ \dot{y} = \frac{\nu}{m} y - \frac{\kappa_0}{m} x $$

# Question 3: What information is missing to complete our ability to solve the ODE system (whether by simulation or analytically)?

# A: We need initiail conditions for position and velocity. This can come in the form of $x(t=0)$ and $\dot{x}(t=0) = y(t=0)$. 

# Question 4: Can you explain the following code meant to simulate the ODE system?

```python
def f(y): 
    return y

def g(x, y, m, nu, k0):
    return -(nu/m)*y - (k0/m)*x

def RK2_step(x,y, m, nu, k0, dt):
    
    # Euler half step
    x_half_step = x + 0.5*dt*f(y)
    y_half_step = y + 0.5*dt*g(x, y, m, nu, k0)
    
    # Runge-Kutta step
    x_new = x + dt*f(y_half_step)
    y_new = y + dt*g(x_half_step, y_half_step, m, nu, k0)
    
    return x_new, y_new

def RK2(x0, y0, m, nu, k0, dt,T):
    
    # Initial data storage
    x_data = [x0]
    y_data = [y0]
    t_data = [0]
    
    while t_data[-1] < T:
        
        # RK2 Step
        x_new, y_new = RK2_step(x_data[-1],y_data[-1], m, nu, k0, dt)
        
        # Storage new data
        x_data.append(x_new)
        y_data.append(y_new)
        t_data.append(t_data[-1]+dt)
        
    return x_data, y_data, t_data
```

```python
x_data, y_data, t_data = RK2(0.5, 0, 1, .01, .1, 0.01,1000)

plt.plot(t_data, x_data)

plt.ylabel('Spring Displacement [m]')
plt.xlabel('Time [s]')
```

```python
def kappa(x, k0,k1,k2):
    return k0 + k1*x + k2*x**2

def nonlinear_force(x,k0,k1,k2):
    return x*kappa(x,k0,k1,k2)
```

# Question 5: Suppose we had the data and knew that the form of the data was 
# $$ f(x) = k_0 + k_1x + k_2x^2 + \varepsilon$$
# where $\varepsilon$ is unbiased normally distributed random noise but did not know the values of the parameters $k_0$, $k_1$, and $k_2$. How might you fit the function 

# $$ f(x) = k_0 + k_1x + k_2x^2$$

# to the data?

```python
def f(y): 
    return y

def g(x, y, m, nu, k0,k1,k2):
    return -(nu/m)*y - (kappa(x,k0,k1,k2)/m)*x

def RK2_step(x,y, m, nu, k0,k1,k2, dt):
    
    # Euler half step
    x_half_step = x + 0.5*dt*f(y)
    y_half_step = y + 0.5*dt*g(x, y, m, nu, k0,k1,k2)
    
    # Runge-Kutta step
    x_new = x + dt*f(y_half_step)
    y_new = y + dt*g(x_half_step, y_half_step, m, nu, k0,k1,k2)
    
    return x_new, y_new

def RK2(x0, y0, m, nu, k0,k1,k2, dt,T):
    
    # Initial data storage
    x_data = [x0]
    y_data = [y0]
    t_data = [0]
    
    while t_data[-1] < T:
        
        # RK2 Step
        x_new, y_new = RK2_step(x_data[-1],y_data[-1], m, nu, k0,k1,k2, dt)
        
        # Storage new data
        x_data.append(x_new)
        y_data.append(y_new)
        t_data.append(t_data[-1]+dt)
        
    return x_data, y_data, t_data
```

# Question 6: What is a norm or metric?

```python
def metric(x_target,x_predicted):
    norm_val = 0
    for i in range(0,len(x_target)):
        norm_val += (x_target[i] - x_predicted[i])**2
        
    norm_val /= len(x_target)
    
    return norm_val
```

```python
metric_data = []
for initial_x in np.linspace(0.1,5,100):
    # nonlinear force simulation 
    x_data_nonlinear, y_data_nonlinear, t_data_nonlinear = RK2(initial_x, 0, 1, .01, k0, k1,k2,0.1,1000)
    
    # linear force simulation 
    x_data_linear, y_data_linear, t_data_linear = RK2(initial_x, 0, 1, .01, k0,  0,0,0.1,1000)
    
    metric_data.append(metric(x_data_nonlinear, x_data_linear))
    
plt.plot(np.linspace(0.1,5,100), metric_data)
plt.xlabel('Initial Value for $x$')
plt.ylabel('Metric Value');
```

# Question 7: Can you sketch out the Gradient Descent Algorithm given a norm which requires one unknown parameter value?

```python
def numerical_derivative(k2):
    
    ### gradient calculation ### 
    delta = 0.001
    
    # plus side of derivative
    simulation_plus = RK2(x0, y0, m, nu, k0,k1,k2+delta, dt,T)
    norm_plus = metric(x_true, simulation_plus[0])
    
    # minus side of derivative 
    simulation_minus = RK2(x0, y0, m, nu, k0,k1,k2-delta, dt,T)
    norm_minus = metric(x_true, simulation_minus[0])
    
    # gradient
    numerical_derivative = (norm_plus - norm_minus) / delta
    
    return numerical_derivative

def GD_step(k2,lamb):
    
    k2 = k2 - lamb*numerical_derivative(k2)
    
    return k2

def GD_method(k2):
    
    lamb = .01
    
    # store initial guess
    k2_data = [k2]
    print(k2_data[-1])
    # generate initial norm value
    simulation = RK2(x0, y0, m, nu, k0,k1,k2, dt,T)
    norm = metric(x_true, simulation[0])
    
    # store initial norm value 
    norm_data = [norm]
    
    # generate derivative data
    num_deriv = numerical_derivative(k2_data[-1])
    
    # store derivative data
    num_deriv_data = [num_deriv]
    
    
    while abs(num_deriv_data[-1]) > 1e-6:
        k2_new = GD_step(k2_data[-1],lamb)
        k2_data.append(k2_new)
        print(k2_data[-1])
        num_deriv = numerical_derivative(k2_data[-1])
        num_deriv_data.append(num_deriv)
        
        simulation = RK2(x0, y0, m, nu, k0,k1,k2_data[-1], dt,T)
        norm = metric(x_true, simulation[0])
        norm_data.append(norm)
    return k2_data, norm_data, num_deriv_data
```
